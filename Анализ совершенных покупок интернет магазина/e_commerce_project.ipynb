### 5. Использовать когортный анализ для выявления когорты с самым высоким Retention Rate на 3 месяц

Retention Rate - это коэффициент, который отражает способность компании выстраивать крепкие отношения с покупателями. Он показывает, насколько хорошо бизнес удерживает постоянных клиентов в течение определенного промежутка времени. Метрику также называют customer retention rate, или CRR.

Рассчитывается она так:
- Retention rate = (покупатели на конец периода - новые покупатели за период) / покупатели на начало периода x 100%

В данном задание повторность покупок является фактором удержания конкретных клиентов.
Нам нужно взять уникальных клиентов, количество покупок, месяца покупок и выбрать только диапазон январь - декабрь. И плюс три месяца покупок для когорты.

Отфильтруем данные для диапазона январь - декабрь

```python
delivered_order['order_purchase_timestamp'].min()
```

```python
delivered_order['order_purchase_timestamp'].max()
```

Для когортного анализа необходимы данные с декобря по январь плюс три месяца для покупок, значит берем диапазон с 2016-12 по 2018-04

```python
delivered_order['year_month'] = delivered_order['order_purchase_timestamp'].dt.to_period("M")
```

```python
delivered_order_range = delivered_order.query('year_month > "2016-12" and year_month < "2018-04"').reset_index()
```

```python
delivered_order_range['year_month'].unique()
```

Для каждого пользователя находим первый месяц покупки

```python
first_buy = delivered_order_range.groupby('customer_unique_id', as_index=False)['year_month'].min()
```

```python
first_buy = first_buy.rename(columns={'year_month': 'firts_month'})
```

```python
delivered_order_range = delivered_order_range.merge(first_buy, how='inner', on='customer_unique_id').reset_index()
```

Находим когорты, группируем по первому месяцу и месяцу покупки и считаем количество пользователей

```python
chogort_df = delivered_order_range.groupby(["firts_month", "year_month"], as_index=False) \
                                  .agg({"customer_unique_id": "nunique"}).sort_values(["firts_month", "year_month"])
```

Считаем разницу между месяцем покупки и первым месяцем

```python
chogort_df['diff'] = (chogort_df['year_month'] - chogort_df['firts_month']).apply(lambda x: x.n)
```

Создаем сводную таблицу, где в строках будет дата первой покупки, столбцами будут временной отрезок, а значением будет количество пользователей купившие продукт

```python
chogort_pivot_df = chogort_df.pivot_table(index='firts_month', columns='diff', \
                                               values='customer_unique_id', aggfunc='sum')
```

Считаем retention rate путем деления количества активных клиентов на конец периода, деленная на количество клиентов на начало периода


```python
cohort_table_percentage = chogort_pivot_df.divide(chogort_pivot_df.iloc[:,0],axis=0)
```

Когорта 2017-06(Июня) является с самым выскоим Retention Rate на 3 месяц

### 6. Провести RFM сегментацию пользователей, для качественной оценки аудитории. Где R - время от последней покупки пользователя до текущей даты, F - суммарное количество покупок у пользователя за всё время, M - сумма покупок за всё время

RFM-анализ — метод анализа, позволяющий сегментировать клиентов по частоте и сумме покупок и выявлять тех клиентов, которые приносят больше денег.

Аббревиатура RFM расшифровывается:

- Recency — давность (как давно ваши пользователи что-то у вас покупали);
- Frequency — частота (как часто они у вас покупают);
- Monetary — деньги (общая сумма покупок).

 Речь идет про покупки, снова берем датафрейм с покупками

Нам потребуется:
- R - время от последней покупки пользователя до текущей даты,
- F - суммарное количество покупок у пользователя за всё время,
- M - сумма покупок за всё время

Находим дату последней покупки для каждого клиента

```python
delivered_order.groupby('customer_unique_id', as_index=False)['order_purchase_timestamp'].max().rename(columns={'order_purchase_timestamp':'last_order'})
```

Дата последней покупки плюс одни день будем считать текущей датой с которой будем сравнивать, как давно клиент что-то покупал

```python
right_now = delivered_order['order_purchase_timestamp'].max() + timedelta(days=1)
```

```python
right_now
```

Добовляем колонку с количеством дней между покупкой и текущей датой

```python
delivered_order['days_between_orders'] = delivered_order['order_purchase_timestamp'].apply(lambda x: (right_now - x).days)
```

Объеденяем нашу таблицу в талбицей товаров и цен на товары

```python
delivered_order_itmes = delivered_order.merge(order_items, on='order_id', how='inner')
```

Создаем фрейм с уникальными id и количество дней спуся прошлой покупки

```python
r = delivered_order_itmes.groupby(['customer_unique_id'], as_index=False)['days_between_orders'].min().rename(columns={'days_between_orders': 'Recency'})
r.head()
```

Создаем фрейм с уникальными id и количеством покупок за все время

```python
f = delivered_order_itmes.groupby(['customer_unique_id'], as_index=False)['order_id'].nunique().rename(columns={'order_id': 'Frequency'})
f.head()
```

Создаем фрейм с уникальными id и суммой потраченных денег за все время

```python
m = delivered_order_itmes.groupby(['customer_unique_id'], as_index=False)['price'].sum().rename(columns={'price': 'Monetary'})
m.head()
```

Объеденим все три таблицы

```python
rfm = r.merge(f, on='customer_unique_id', how='inner').merge(m, on='customer_unique_id', how='inner')
rfm.head()
```

```python
rfm.describe()
```

Минимум всех покупателей совершили одну покупку. Для показателя Frequency будет только 2 кластера: одна покупка и более одной покупки

```python
rfm.Frequency.value_counts(normalize=True).multiply(100).round(3)
```

Recency и Monetary распределим по квинтилям, в каждом из квинтилей по 20% от наблюдений

```python
quintiles = rfm[['Recency', 'Monetary']].quantile([0.2, 0.4, 0.6, 0.8]).to_dict()
quintiles
```

Считаем методы для присвоения рангов по показателям Recency, Frequency и Monetary

```python
def r_score(x):
    if x <= quintiles['Recency'][0.2]:
        return 5
    elif x <= quintiles['Recency'][0.4]:
        return 4
    elif x <= quintiles['Recency'][0.6]:
        return 3
    elif x <= quintiles['Recency'][0.8]:
        return 2
    else:
        return 1

def m_score(x):
    if x <= quintiles['Monetary'][0.2]:
        return 1
    elif x <= quintiles['Monetary'][0.4]:
        return 2
    elif x <= quintiles['Monetary'][0.6]:
        return 3
    elif x <= quintiles['Monetary'][0.8]:
        return 4
    else:
        return 5

def f_score(x):
    if x == 1:
        return 1
    else:
        return 2
```

Получаем оценки R, F и M каждого клиента.

```python
rfm['R'] = rfm['Recency'].apply(lambda x: r_score(x))
rfm['F'] = rfm['Frequency'].apply(lambda x: f_score(x))
rfm['M'] = rfm['Monetary'].apply(lambda x: m_score(x))
```

```python
rfm.head()
```

Объединяем оценки R, F и M в один балл RFM Score

```python
rfm['RFM_Score'] = rfm['R'].map(str) + rfm['F'].map(str) + rfm['M'].map(str)
```

```python
rfm.head()
```

```python
rfm['R'].map(str) + rfm['M'].map(str)
```

Описываем названия сегментов

```python
segt_map = {
    r'[1-2][1-2]': 'спящие',
    r'[1-2][3-4]': 'могут уснуть',
    r'[1-2]5': 'требуют внимания',
    r'3[1-2]': 'почти спящие',
    r'33': 'нужно обратить внимание',
    r'[3-4][4-5]': 'лояльные клиенты',
    r'41': 'многообещающие',
    r'51': 'новые клиенты',
    r'[4-5][2-3]': 'потенциально лояльные',
    r'5[4-5]': 'очень лояльные'
}

rfm['segment'] = rfm['R'].map(str) + rfm['M'].map(str)
rfm['segment'] = rfm['segment'].replace(segt_map, regex=True)
rfm.head()
```

```python
rfm_end = rfm.groupby('segment', as_index=False)['customer_unique_id'].count() \
                                                            .rename(columns={'customer_unique_id':'count_customers'}) \
                                                            .sort_values('count_customers', ascending=False).reset_index(drop=True)
rfm_end['percentage'] = round((rfm_end['count_customers'] / delivered_order['customer_unique_id'].nunique()) * 100).astype(str) + '%'
rfm_end
```
